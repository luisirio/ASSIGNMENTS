```python
raw_data = pd.read_csv(f"{os.path.dirname(os.getcwd())}/life_expectancy/data/eu_life_expectancy_raw.tsv", sep='\t')
```

```python
raw_data.head()
```

```python
initial_columns = raw_data.columns[0]
```

```python
initial_columns
```

```python
new_columns = raw_data.columns[0].split(',')
```

```python
new_columns
```

```python
raw_data[new_columns] = raw_data[initial_columns].str.split(',', expand=True)
```

```python
raw_data.head()
```

```python
raw_data = raw_data.drop(columns=initial_columns)
```

```python
raw_data.head()
```

```python
melted_df = pd.melt(
            raw_data,
            new_columns,
            var_name='year',
            value_name='value'
            )
```

```python
melted_df.head(20)
```

```python
melted_df.columns
```

```python
print('geo\\time')
```

```python
melted_df.rename(columns={'geo\\time':'region'}, inplace = True)
```

```python
print(melted_df.head())
```

```python
melted_df['year'] = melted_df['year'].astype('int')
```

```python
melted_df['year'].dtypes
```

```python
# remove non-numeric characters and extract float values
melted_df['value'] = melted_df['value'].str.extract('(\d+\.\d+)', expand=False).astype(float)
```

```python
print(melted_df.head(100))
```

```python
melted_df = melted_df[melted_df['region']== 'PT']
```

```python
# drop NaN values
melted_df = melted_df.dropna(subset=['value'])
```

```python
melted_df.to_csv(f"{os.path.dirname(os.getcwd())}/life_expectancy/data/pt_life_expectancy.csv", index = False)
```
